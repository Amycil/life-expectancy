```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

#load and expect data
life_expectancy_gdp = pd.read_csv('all_data.csv')
life_expectancy_gdp.head()
```

```python
#EDA
life_expectancy_gdp.isnull().sum()

life_expectancy_gdp.describe()
```

```python
life_expectancy_gdp.Country.unique()

life_expectancy_gdp.Year.unique()

#rename life expectancy column
life_expectancy_gdp = life_expectancy_gdp.rename({'Life expectancy at birth (years)':
'Life_expectancy'}, axis=1)
```

```python
#summarize table - group by means
new_df_means = life_expectancy_gdp.drop('Year', axis=1).groupby('Country').mean().reset_index()
new_df_means
```

```python
#plot life_expectancy vs country

sns.barplot(x='Life_expectancy', y='Country', data= new_df_means)
plt.title("Life Expectancy per country")
plt.show()
```

```python
#gdp vs life expectancy
sns.scatterplot(data=life_expectancy_gdp, x='Life expectancy', y='GDP', hue='Country')
plt.show()
```

```python
g = sns.FacetGrid(data=life_expectancy_gdp, col='Country', hue='Country', col_wrap=3, sharey=False, sharex=False)
g.map(sns.scatterplot, 'Life expectancy', 'GDP').add_legend()
plt.show()
```

```python
#gdp over time
g = sns.FacetGrid(data=life_expectancy_gdp, col='Country', col_wrap=3, hue='Country', sharey=False, sharex=False)
g.map(sns.lineplot, 'Year', 'GDP').add_legend()
plt.show()
```

```python
#life expectancy and year

g = sns.FacetGrid(life_expectancy_gdp, col='Country', hue='Country', col_wrap=3, sharey=False, sharex=False)
g.map(sns.lineplot, 'Year', 'Life_expectancy').add_legend()
plt.show()
```

## TAKEAWAY FROM VISUALIZATIONS
-  There is a positive linear relationship between `GDP` and `Life Expectancy at Birth`. The countries experienced an increase in life expectancy over the years as GDP increased.
- Even though all the countries saw an increase in GDP over time, Zimbabwe seemed to have had a steadier rise in GDP and at a much faster rate.
